# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
x = dataset.iloc[:, :-1].values #takes till before last column
y = dataset.iloc[:, -1].values
```

```
print(x)
```

```
print(y)
```

```
z=dataset.iloc[:, 0:-1].values
```

```
print(z)
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan , strategy = 'mean') #np is numpy
imputer.fit(x[:, 1:3])
x[:, 1:3]=imputer.transform(x[:, 1:3])
```

```
print(x)
```

## Encoding categorical data

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers= [('encoder',OneHotEncoder(),[0])],remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```
print(x) # 1 0 0 for france     0 0 1 for spain     0 1 0 for germany
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(y)
```
